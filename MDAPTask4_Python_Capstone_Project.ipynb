```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
l = ['Hotel_Address', 'Additional_Number_of_Scoring', 'Review_Date', 'Average_Score', 'Hotel_Name', 'Reviewer_Nationality', 'Negative_Review', 'Review_Total_Negative_Word_Counts', 'Review_Total_Positive_Word_Counts','Positive_Review', 'Reviewer_Score', 'Total_Number_of_Reviews_Reviewer_Has_Given', 'Total_Number_of_Reviews', 'Tags', 'days_since_review', 'lat', 'lng']
```

```python
# importing datafile

df = pd.read_csv('Hotel_reviews.csv', header = 0, names = l)
df
```

```python
df.columns
```

```python
df.shape
```

```python
df.info()
```

```python
df['Positive_Review'] = df['Positive_Review'].replace(['', ' ', np.nan],'No Positive')
df = df[(df['Reviewer_Score'] >=0) & (df['Review_Total_Negative_Word_Counts'] >=0) & (df['Review_Total_Positive_Word_Counts'] >=0)]
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace = True)
```

```python
df.info()
```

```python
df.drop('lat', axis = 1, inplace = True)
df.drop('lng', axis = 1, inplace = True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop(columns = 'Tags', inplace = True)
```

```python
df.isnull().sum()
```

```python
num_df = df.select_dtypes(include = np.number)
num_df.head()
```

```python
num_df.corr()
```

```python
plt.figure(figsize = (12,12))
sns.heatmap(num_df.corr(), annot = True)
plt.show()
```

Additional_Number_of_Scoring is directly proportional to Review_Total_Positive_Word_Counts, whereas Average_Score is inversely proportional to Review_Total_Positive_Word_Counts.

```python
# Top hostels according to Reviewer Score
plt.figure(figsize = (8,5))
sns.countplot(x = pd.Series(df['Hotel_Name'].unique()).sort_values(ascending = False).iloc[1:10])
plt.title('Top 10 Hotels with Highest Reviewer Scores')
plt.xticks(rotation = 90)
plt.show()
```

```python
# Top worst hostels according to Reviewer Score
plt.figure(figsize = (8,5))
sns.countplot(x = pd.Series(df['Hotel_Name'].unique()).sort_values(ascending = True).iloc[1:10])
plt.title('Top 10 Hotels with Lowest Reviewer Scores')
plt.xticks(rotation = 90)
plt.show()
```
